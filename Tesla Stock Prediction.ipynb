# Tesla Stock Price Prediction

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import preprocessing
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from matplotlib import style
style.use("fivethirtyeight")
```

```python
df= pd.read_csv('Tesla.csv+-+Tesla.csv.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["HL_Perc"] = (df["High"]-df["Low"]) / df["Low"] * 100
df["CO_Perc"] = (df["Close"] - df["Open"]) / df["Open"] * 100
```

```python
dates = np.array(df["Date"])
dates_check = dates[-30:]
dates = dates[:-30]
```

```python
df.head()
```

```python
dates = np.array(df["Date"])
```

```python
df = df[["HL_Perc", "CO_Perc", "Adj Close", "Volume"]]
```

```python
df["PriceNextMonth"] = df["Adj Close"].shift(-30)
```

```python
df.tail()
```

```python
X = np.array(df.drop(["PriceNextMonth"],axis=1))
X = preprocessing.scale(X)
X_Check = X[-30:]
X = X[:-30]
df.dropna(inplace = True)
y = np.array(df["PriceNextMonth"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
conf = model.score(X_test, y_test)
print(conf)
```
